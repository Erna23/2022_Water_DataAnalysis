# 데이터 전처리
- 측정시간 -> 측정일(월/일)
- 대체 못한 결측치 제외
- 측정일 별로 그룹지어 각 변수 평균값 구함

```python
df['측정시간'] = df['측정시간'].astype('str')
```

```python
df.info()
```

```python
df['월'] = df['측정시간'].str.split('-',expand=True)[1]
df['일'] = df['측정시간'].str.split('-',expand=True)[2]
df['일'] = df['일'].str.split(' ',expand=True)[0]
df.head()
```

```python
df = df.drop(['측정시간'],axis=1)
df.head()
```

```python
df['측정일'] = df['월']+"/"+df['일']
df.tail()
```

```python
df = df.drop(['월','일'],axis=1)
df.head()
```

```python
df[df['pH']=='결측'].index
df = df.drop([331,517],axis=0)
```

```python
df.set_index('측정일',inplace=True)
df = df.astype('float')
df.reset_index(inplace=True)
```
