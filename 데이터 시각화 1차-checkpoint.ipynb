```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib 
matplotlib.rcParams['font.family'] = 'Malgun Gothic' 
matplotlib.rcParams['font.size'] = 10 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False

import datetime as dt
import time as tm
```

```python
df= pd.read_excel('솔정.xlsx')
df.head()
```

```python
df=df.drop(['비고'],axis=1)
df
```

```python
df.isnull().sum()
```

```python
df['측정시간'] = df['측정시간'].astype('str')
```

```python
df.info()
```

```python
df['월'] = df['측정시간'].str.split('-',expand=True)[1]
df['일'] = df['측정시간'].str.split('-',expand=True)[2]
df['일'] = df['일'].str.split(' ',expand=True)[0]
df.head()
```

```python
df = df.drop(['측정시간'],axis=1)
df.head()
```

```python
df['측정일'] = df['월']+"/"+df['일']
df.tail()
```

```python
df_2 = df.groupby(df['월']).mean()
df_2
```

## 데이터를 월 단위로 시각화하였을때

- TOC는 T-N, 수위(m)에 대체로 반비례함.
- TOC는 T-P에 대체로 비례함.

```python
df = df.drop(['월','일'],axis=1)
df.set_index('측정일',inplace=True)
df = df.astype('float')
df.reset_index(inplace=True)
```

```python
df_1 = df.groupby(df['측정일']).mean()
df_1
```

```python
df_1['TOC(mg/L)'].sort_values()
df_1['T-N(mg/L)'].sort_values()

df_1['TOC(mg/L)'].plot(color='red')
df_1['T-N(mg/L)'].plot(color='purple')

plt.ylim(0,10)
plt.show()
```

```python
df_1['TOC(mg/L)'].sort_values()
df_1['수위(m)'].sort_values()

df_1['TOC(mg/L)'].plot(color='red')
df_1['수위(m)'].plot(color='yellow')

plt.ylim(0,3)
```

```python
df_1['T-P(mg/L)'].plot(color='blue')
```

## 데이터를 일 단위로 그래프를 시각화했을때

- 7월 : TOC는 T-N에 대체로 반비례함, TOC는 T_P와 수위(m)에 대체로 비례함.
- 8월~12월 : TOC는 T_N과 T-P에 대체로 비례함, TOC는 수위(m)에 대체로 반비례함.
