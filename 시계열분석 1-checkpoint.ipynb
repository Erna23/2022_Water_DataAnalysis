```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib 
matplotlib.rcParams['font.family'] = 'Malgun Gothic' 
matplotlib.rcParams['font.size'] = 10 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False

import datetime as dt
import time as tm
```

```python
df= pd.read_excel('솔정2.xlsx')
df
```

```python
df['측정시간'] = pd.to_datetime(df['측정시간'])
df
```

```python
df.index = df['측정시간']
df.set_index('측정시간', inplace=True)
```

```python
df.plot()
plt.show()
```

```python
df = df.asfreq('H', method='ffill')
```

```python
df.plot(kind='line', figsize=(20,6), linewidth=3, fontsize=20, ylim=(0,5)) 
plt.title('Time Series of Target', fontsize=20) 
plt.xlabel('MONTH', fontsize=15) 
plt.ylabel('TOC', fontsize=15) 
plt.show()
```

## 이동평균 계산


```python
# comparison of several moving average values
pd.concat([df[['TOC']],#시간정보(파랑)
           df[['TOC']].rolling(24).mean(), #데일리 패턴을 보겠다.(주황색)
           df[['TOC']].rolling(24*7).mean()], axis=1).plot(kind='line', figsize=(20,6), linewidth=3, fontsize=20, ylim=(0,5))
plt.title('Time Series of Target', fontsize=20) # 주별패턴을 보겠다(녹색).
plt.xlabel('MONTH', fontsize=15) 
plt.ylabel('TOC', fontsize=15) 
plt.show()
```

```python
# fill nan as some values and merging
Y_TOC_Day = df[['TOC']].rolling(24).mean()
Y_TOC_Day.fillna(method='ffill', inplace=True) #24개를 합쳤기 때문에 nan값이 있을거임 이를 채움
Y_TOC_Day.fillna(method='bfill', inplace=True)
Y_TOC_Day.columns = ['TOC_Day']
Y_TOC_Week = df[['TOC']].rolling(24*7).mean()
Y_TOC_Week.fillna(method='ffill', inplace=True)
Y_TOC_Week.fillna(method='bfill', inplace=True)
Y_TOC_Week.columns = ['TOC_Week']
df = pd.concat([df, Y_TOC_Day], axis=1)
df = pd.concat([df, Y_TOC_Week], axis=1)
df
```

## 증감폭 계산

```python
# line plot of Y for specific periods 바뀌는 차이를 알고싶을때 diff()를 사용, 증감폭을 알 수 있음
df[['TOC']].diff().plot(kind='line', figsize=(20,6), linewidth=3, fontsize=20,
                                 ylim=(-5,5))
plt.title('Time Series of Target', fontsize=20)
plt.xlabel('MONTH', fontsize=15) 
plt.ylabel('TOC', fontsize=15) 
plt.show()
```

```python
# diff of Y and merging
Y_diff = df[['TOC']].diff()
Y_diff.fillna(method='ffill', inplace=True)
Y_diff.fillna(method='bfill', inplace=True)
Y_diff.columns = ['TOC_diff']
df = pd.concat([df, Y_diff], axis=1)
df
```
